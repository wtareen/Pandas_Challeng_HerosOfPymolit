* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results

```python
purchase_data.head()
```

```python
purchasing_Analysis1 = purchase_data.groupby(['Gender']).agg({'Purchase ID':'count','Price':['mean','sum']}).droplevel(0,axis = 1) \
.rename(columns = {"count":"Purchase Count", "mean": "Average Purchase Price", "sum":"Total Purchase Value"})
purchasing_Analysis1
```

```python
purchasing_Analysis2 = purchase_data.groupby(['Gender','SN']).agg({'Price':'sum'}).reset_index().groupby(['Gender']).agg({'Price':'mean'}) \
.rename(columns ={'Price': 'Average Purchase Total'})
purchasing_Analysis2
```

# Answer - Purchasing Analysis (Gender)

```python
purchasing_Analysis = pd.concat([purchasing_Analysis1,purchasing_Analysis2],axis = 1)
f5 = lambda x : '${0:.2f}'.format(x)
purchasing_Analysis['Average Purchase Price'] = purchasing_Analysis['Average Purchase Price'].apply(f5)
purchasing_Analysis['Average Purchase Total'] = purchasing_Analysis['Average Purchase Total'].apply(f5)
purchasing_Analysis['Total Purchase Value'] = purchasing_Analysis['Total Purchase Value'].apply(f5)
purchasing_Analysis
```

# Answer Most Popular Itmes
